### filter() function

```python
# Returning all number greater than 5 from a list of numbers
numbers = [4,3,6,7,9,3,11]

def filt_grt5(num):
    return num > 5
```

```python
filt = filter(filt_grt5, numbers)
list(filt)
```

```python
# Can also be inline filter using a Lambda function
filt = filter(lambda x:x > 5,numbers)
list(filt)
```
